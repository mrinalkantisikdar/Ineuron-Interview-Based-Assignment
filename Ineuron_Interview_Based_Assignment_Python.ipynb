## Question 1: -
Write a program that takes a string as input, and counts the frequency of each word in the string, there might
be repeated characters in the string. Your task is to find the highest frequency and returns the length of the
highest-frequency word.
Note - You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.
Example input - string = “write write write all the number from from from 1 to 100”
Example output - 5
Explanation - From the given string we can note that the most frequent words are “write” and “from” and
the maximum value of both the values is “write” and its corresponding length is 5


```python
input_string = "write write write all the number from from from 1 to 100"

def word_count(str):
    counts = dict()
    words = str.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1

    counts_x =dict(sorted(counts.items(), key= lambda key_value: key_value[1])) # sorts key_values in ascending order & convert to dictionary
    return counts_x

my_dict=word_count(input_string)
my_dict
```

```python
# finding duplicate values
# from dictionary using flip
def flip_dict(dict):
  flipped = {}
  
  for key, value in dict.items():
      if value not in flipped:
          flipped[value] = [key]
      else:
          flipped[value].append(key)
  return flipped

new_dict=flip_dict(my_dict)
new_dict
```

```python
last_key=list(new_dict)[-1] 
# last key = list(my dict)[-1]. The dictionary is transformed into a list of keys by the list class, and the last key may be obtained by accessing the element at index -1
last_key
```

```python
my_list= new_dict.get(last_key)
# get()	Returns the value of the specified key
my_list
```

```python
max(my_list, key=len)
```

```python
length=len(max(my_list, key=len))
length
```

```python
def get_length_most_freq(dict):
  last_key=list(dict)[-1]
  my_list= dict.get(last_key)
  length=len(max(my_list, key=len))
  return length
```

```python
get_length_most_freq(new_dict)
```

```python
str_20= "the long long long brown fox fox fox jumps over the lazy dog"
dict_21= word_count(str_20)
dict_22= flip_dict(dict_21)
len_23= get_length_most_freq(dict_22)
len_23
```

Here both the words long and fox appears 3 times. But the word long is longer with 4 characters.

```python
str_30= "it it it is a beautiful day day day"
dict_31= word_count(str_30)
dict_32= flip_dict(dict_31)
len_33= get_length_most_freq(dict_32)
len_33
```

## Question 2: -
Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if
he can remove just one character at the index in the string, and the remaining characters will occur the same
number of times. Given a string, determine if it is valid. If so, return YES , otherwise return NO .
Note - You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.
Example input 1 - s = “abc”. This is a valid string because frequencies are { “a”: 1, “b”: 1, “c”: 1 }
Example output 1- YES
Example input 2 - s “abcc”. This string is not valid as we can remove only 1 occurrence of “c”. That leaves
character frequencies of { “a”: 1, “b”: 1 , “c”: 2 }
Example output 2 - NO

```python

def string_freq(str):
  freq = {}
  inp_str= str  
  for ele in inp_str:
      if ele in freq:
          freq[ele] += 1
      else:
          freq[ele] = 1
  return freq
```

```python
s1= 'abc'
d1=string_freq(s1)
d1
```

```python
def validate(dict):
  myset= set(dict.values()) # taking the dictionary values and making a set of it to remove the duplicates
  if len(myset)==1:
    for item in myset:
      if item==1: 
        return 'YES'
  else:
    return 'NO'
```

```python
validate(d1)
```

The string s1 doesn't have any repeating characters, so the result is YES.

```python
s2= 'abcc'
d2= string_freq(s2)
d2
```

```python
validate(d2)
```

The string s2 has repeating characters, so the result is NO.

```python
s3= 'abcdef'
d3= string_freq(s3)
validate(d3)
```

The string s3 doesn't have any repeating characters, so the result is YES.

```python
s4= 'aabcdef'
d4= string_freq(s4)
validate(d4)
```

The string s4 has repeating characters, so the result is NO.
